```
%matplotlib inline
import numpy as np
import plotly 

import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 15, 10

plt.style.use('ggplot')

x=np.arange(-100,5,0.01)
plt.plot(x,3-np.log(x)+x*x*x+90*np.sin(x)*np.cos(5*x)-3*np.cos(3*x))

plt.title('Stack Plot Example')
plt.legend()
plt.show()
```
